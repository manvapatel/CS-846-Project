```python
#To see the labels on this barplot----

groupedvalues = data_subset2.groupby(['contrib_country','contrib_gender']).agg(merged_fraction=("merged_or_not",'mean'))
 
```

```python
groupedvalues=groupedvalues.reset_index()
```

```python
groupedvalues
```

```python
import matplotlib.pyplot as plt
```

```python
#plot barplot for male dominated regions around asia & africa
plt.figure(figsize=(10, 8))
splot=sns.barplot(x="contrib_country", y="merged_fraction", hue="contrib_gender", data=groupedvalues.iloc[[8,9,10,11,18,19,35,36,37,38], :])
plt.ylabel("Merged requests fraction", size=14)
plt.xlabel("Region of the contributor", size=14)
plt.xticks(rotation=90)
for p in splot.patches:
    splot.annotate(format(p.get_height(),'0.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=15,
                   xytext = (0, -12), 
                   textcoords = 'offset points')
```

```python
# finding countries where fraction of requests merged among those put by males > females scenario
region_count = groupedvalues.groupby(['contrib_country','contrib_gender']).mean().squeeze().unstack().add_suffix('_mergecount')
print (region_count)
```

```python
m_regions= region_count[region_count["female_mergecount"] < region_count["male_mergecount"]]
print(m_regions)
```

```python
f_regions= region_count[region_count["female_mergecount"] > region_count["male_mergecount"]]
print(f_regions)
```
