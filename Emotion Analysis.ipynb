```python
#data_mean = data_subset1.groupby(['contrib_country', 'contrib_gender'], as_index = False)['perc_pos_emotion', 'perc_neg_emotion', 'perc_neu_emotion'].mean()
data_grouped = data_subset1.groupby(['contrib_country', 'contrib_gender'], as_index = False).agg({'project_id':'size', 'perc_pos_emotion':'mean', 'perc_neg_emotion':'mean','perc_neu_emotion':'mean'}) \
       .rename(columns={'project_id':'count'}) 
data_grouped
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15, 12))
splot=sns.barplot(x="contrib_country", y="perc_pos_emotion", hue="contrib_gender", data= data_grouped)
plt.ylabel("Merged requests fraction", size=14)
plt.xlabel("Region of the contributor", size=14)
plt.xticks(rotation=90)
```
