```python
data_subset2 = data[['project_id', 'ownername', 'reponame','merged_or_not', 'contrib_gender', 'contrib_country', 'inte_country', 'same_country']].dropna(subset=['contrib_gender', 'contrib_country'])
print(data_subset2.shape)
data_subset2.head()

```

```python
data_subset2.loc[data_subset2["contrib_country"].isin(sub_saharan_africa), "contrib_country"] = "sub_saharan_africa"
data_subset2.loc[data_subset2["contrib_country"].isin(northern_africa), "contrib_country"] = "northern_africa"
data_subset2.loc[data_subset2["contrib_country"].isin(western_asia), "contrib_country"] = "western_asia"
data_subset2.loc[data_subset2["contrib_country"].isin(central_asia), "contrib_country"] = "central_asia"
data_subset2.loc[data_subset2["contrib_country"].isin(southern_asia), "contrib_country"] = "southern_asia"
data_subset2.loc[data_subset2["contrib_country"].isin(eastern_asia), "contrib_country"] = "eastern_asia"
data_subset2.loc[data_subset2["contrib_country"].isin(south_eastern_asia), "contrib_country"] = "south_eastern_asia"
data_subset2.loc[data_subset2["contrib_country"].isin(caribbean), "contrib_country"] = "caribbean"
data_subset2.loc[data_subset2["contrib_country"].isin(central_america), "contrib_country"] = "central_america"
data_subset2.loc[data_subset2["contrib_country"].isin(south_america), "contrib_country"] = "south_america"
data_subset2.loc[data_subset2["contrib_country"].isin(australia_and_new_zealand), "contrib_country"] = "australia_and_new_zealand"
data_subset2.loc[data_subset2["contrib_country"].isin(melanesia), "contrib_country"] = "melanesia"
data_subset2.loc[data_subset2["contrib_country"].isin(micronesia), "contrib_country"] = "micronesia"
data_subset2.loc[data_subset2["contrib_country"].isin(polynesia), "contrib_country"] = "polynesia"
data_subset2.loc[data_subset2["contrib_country"].isin(northern_america), "contrib_country"] = "northern_america"
data_subset2.loc[data_subset2["contrib_country"].isin(eastern_europe), "contrib_country"] = "eastern_europe"
data_subset2.loc[data_subset2["contrib_country"].isin(western_europe), "contrib_country"] = "western_europe"
data_subset2.loc[data_subset2["contrib_country"].isin(northern_europe), "contrib_country"] = "northern_europe"
data_subset2.loc[data_subset2["contrib_country"].isin(southern_europe), "contrib_country"] = "southern_europe"
data_subset2.loc[data_subset2["contrib_country"].isin(["mozambique"]), "contrib_country"] = "sub_saharan_africa"
```

```python
#To see the labels on this barplot----

groupedvalues = data_subset2.groupby(['contrib_country','contrib_gender']).agg(merged_ratio=("merged_or_not",'mean'))
 
```

```python
groupedvalues=groupedvalues.reset_index()
```

```python
groupedvalues
```

```python
import matplotlib.pyplot as plt
```

```python
#plot barplot for regions belonging to first 11 rows of grouped_values
plt.figure(figsize=(10, 8))
splot=sns.barplot(x="contrib_country", y="merged_ratio", hue="contrib_gender", data=groupedvalues.iloc[0:12, :])
plt.ylabel("Merged requests ratio", size=14)
plt.xlabel("Region of the contributor", size=14)
plt.xticks(rotation=90)
#plt.title("Annotated Grouped Barplot: Gender Bias in Salary", size=18)
for p in splot.patches:
    splot.annotate(format(p.get_height(),'0.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=15,
                   xytext = (0, -12), 
                   textcoords = 'offset points')
#plt.savefig('Annotating_Bars_in_Grouped_Barplot_Seaborn_Matplotlib_try1.png',dpi=150)
```
